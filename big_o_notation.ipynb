```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
```

```python
x = range(1, 9)
```

```python
# y=0n+1 and label the curve "O(1)"
y1 = [(0 * n + 1) for n in x ]
# y=log(n) and label the curve "O(log n)"
y2 = [np.log(x) for x in x]
# y=n and label the curve "O(n)"
y3 = [n for n in x]
# y=nâˆ—log(n) and label it "O(n log n)"
y4 = [n * np.log(n) for n in x]
# y=n2 and label it "O(n^2)"
y5 = [(n ** n) for n in x ]
# y=2n and label it "O(2^n)"
y6 = [(2 ** n) for n in x]
# y=n! and label it "O(n!)"
y7 = [np.math.factorial(n) for n in x]
# y=nn and label it "O(n^n)"
y8 = [(n ** n) for n in x]
```

```python
plt.plot(y1)
plt.plot(y2)
plt.plot(y3)
plt.plot(y4)
plt.plot(y5)
plt.plot(y6)
plt.plot(y7)
plt.plot(y8)
plt.axis([0, 9, 0, 90])
```
