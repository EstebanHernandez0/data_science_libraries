```python
import numpy as np
import matplotlib.pyplot as plt
import math as math
%matplotlib inline
```

```python
v = range(-100, 100)
```

```python
# Use matplotlib to plot the following equation:
y = [(num ** 2 - num + 2) for num in v]
```

```python
plt.plot(v, y)
plt.title('$ y = x^2 - x + 2$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.annotate('Origin', xy = (0,0), xytext= (0,3000), arrowprops = {})
plt.ylim(0,10500)
```
